### Introduction to MLOps and Version Control

As the class practice, the students will be required to carry out the following activity:

1. Create a github account with the following user name format: `{name_first_letter}{surname}-at-{institution}`.

2. Configure git with github to work locally if not configured yet.

3. Create a public repository called `mlops-and-system-design`. The repository must include the `.gitignore` file (use template provided by GitHub).

4. Clone the `mlops-and-system-design` to your local.

5. Open the project in Visual Studio Code (please make sure you are on the correct directory).

6. In the root folder, create a `requirements.txt` file. It must include the following:

numpy==1.26.4

joblib==1.4.2

scikit-learn==1.6.1

pandas==2.2.3

ipykernel==6.29.5

pytest==8.3.5

autopep8

isort

black==25.1.0

6. Commit the changes and push them to the remote (`main` branch).

`git add <filename>`

`git commit -m '<message>'`

`git push origin <remote branch>`

7. Check that the changes applied to the remote (Github UI).

8. In the Github UI, create a `main.py` file with the following code in a folder called `session_1` (changes need to be commited to the `main` branch):

```python
def hello_world():
    print("Hello world")
```

9. On the local repositoty, pull the changes from the remote.

`git pull origin <remote branch>`

10. Create a feature branch (`feat/update-hello-world`) to update the function to have a name as input. The output of the function should be like *Hello world, my name is \<NAME_INPUT> and I'm learning about version control!*

*Tip: use git checkout -b <branch_name> to create and switch to a new branch*

```python
hello_world(name="John Smith")  # Hello world, my name is John Smith and I'm learning about version control!
```

11. Commit and  push changes to the remote using this command: `git push origin feat/update-hello-world`

*Note: keep in mind we need to push the changes to the new created branch, not to the `main` branch*

12. Go the Github UI and open a pull request (PR) from the feature branch into the main branch.

13. Merge the changes into the main branch.

14. Remove the feature branch from your local and remote.

Remember to use appropiate naming convention for your branches and commits

**Each student will need to share the repository with the professor, as this will be taken into consideration for the evaluation**
